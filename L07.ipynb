## Task 1b: Load Data

Load the Iris data (same as in the Pandas tasks).


```python
iris = pd.read_csv('data/iris.csv')
iris.head(10)
```

## Task 2a: Create a figure

Create the a scatter plot using the iris data, but compare `petal_width` to `petal_length`.  Be sure to add labels to the axes.

```python
## Create the scatterplot. If using coordinates from different data sets, this would be good. must be same lengths though.


fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.scatter(iris['petal_length'], iris['petal_width'], c = 'r')


# Add axes labels
ax.axes.set_xlabel('Petal Length')
ax.axes.set_ylabel('Petal Width')

# Add a title
ax.set_title('Petal length vs width')

# Show the plot
plt.show()
```

## Task 3a: Figure Style

Create a scatter plots using the iris data frame. Do not convert the data frame columns to numpy arrays. Rather use the dataframe as is. Be sure to:
- plot `petal_length` vs `petal_width`.
- Use column names to specify the x and y coordinates (don't pass in Numpy arrays)
- Color each dot by the species (you'll need to be creative to do this).
  - Hint 1: use the [pd.Series.replace](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.replace.html) function to convert species names to colors). 
  - Hint 2: Use the color abbreviations we learned for the the plot function (e.g. 'r', 'b', 'g', etc.).
- Add axis labels

```python
#Make a new column identical to the species column that is named color
iris['color'] = iris['species']    
iris['color'].replace('setosa', 'r', inplace=True)
iris['color'].replace('versicolor', 'g', inplace=True)
iris['color'].replace('virginica', 'b', inplace=True)

iris.sample(10)

# Create the plot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
nice_plot = ax.scatter('petal_length', 'petal_width', c = 'color', data = iris)


# Add axes labels
ax.axes.set_xlabel('Petal Length')
ax.axes.set_ylabel('Petal Width')

# Add a title
ax.set_title('Petal length vs width')

# Show the plot
plt.show()
```

## Task 4a: Categorical plots

Create a bar graph that indicates the mean `sepal_length` of each subspecies.

```python
means = iris.groupby(by='species', axis=0).mean()
means

plt.bar(means.index, means['sepal_length'].values)
```

## Task 6a:  Multiple Plots

Create 4 scatterplots for the iris data all in one figure. The figure should include scatterplots of these comparisions:
- `sepal_length` x `petal_length`
- `sepal_length` x `petal_width`
- `sepal_width` x `petal_length`
- `sepal_width` x `petal_width`

Be sure that:
- the points for the species 
- the axes for each plot are labeled
- the figure has a label.

```python
# First plot
fig = plt.figure()



sub1 = fig.add_subplot(2, 2, 1)
sub1.scatter('petal_length', 'petal_width', c = 'color', data = iris)


# Add axes labels
sub1.axes.set_xlabel('Petal Length')
sub1.axes.set_ylabel('Petal Width')


# Add a title
sub1.set_title('sepal_length x petal_length')




## Second plot
sub2 = fig.add_subplot(2, 2, 2)
sub2.scatter('sepal_length', 'petal_length', c = 'color', data = iris)
sub2.axes.set_xlabel('Sepal Length')
sub2.axes.set_ylabel('Petal length')
sub2.set_title('sepal_length x petal_width')


## Third plot
sub3 = fig.add_subplot(2, 2, 3)
sub3.scatter('sepal_width', 'petal_length', c = 'color', data = iris)
sub3.axes.set_xlabel('Sepal width')
sub3.axes.set_ylabel('Petal length')
sub3.set_title('sepal_width x petal_length')




## Fourth plot
sub4 = fig.add_subplot(2, 2, 4)
sub4.scatter('sepal_width', 'petal_width', c = 'color', data = iris)
sub4.axes.set_xlabel('Petal width')
sub4.axes.set_ylabel('Petal width')
sub4.set_title('sepal_width x petal_width')




# Add a title to the figure.
fig.suptitle('Sepal and Petal plots')

#Fix the spacing
fig.tight_layout(pad=2)

plt.show()
```

```python
# More efficent way

fig = plt.figure()


plot1 = plt.subplot(2, 2, 1)
plot2 = plt.subplot(2, 2, 2)
plot3 = plt.subplot(2, 2, 3)
plot4 = plt.subplot(2, 2, 4)


plot1.plot(iris['sepal_length'], iris['petal_length'], 'r*')
plot2.plot(iris['sepal_length'], iris['petal_width'], 'g+')
plot3.plot(iris['sepal_width'], iris['petal_length'], 'b--')
plot4.plot(iris['sepal_width'], iris['petal_width'], 'y>')



plot1.set_title('Length: Sepal vs. Petal')
plot2.set_title('Sep length vs. Pet width')
plot3.set_title('Sep width vs. Pet length')
plot4.set_title('Width: Sepal vs. Petal')


plot1.set_xlabel('Sepal Length')
plot2.set_xlabel('Sepal Length')
plot3.set_xlabel('Sepal Width')
plot4.set_xlabel('Sepal Width')


plot1.set_ylabel('Petal Lenth')
plot2.set_ylabel('Petal Width')
plot3.set_ylabel('Petal length')
plot4.set_ylabel('Petal Width')




# Add a title to the figure.
fig.suptitle('Sepal and Petal plots')

fig.tight_layout(pad=2.5)
plt.show()
```
